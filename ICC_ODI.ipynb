```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
odi_bat=pd.read_csv("D:\Practical\Data Sets for Practics\Cricket\Odi_bat.csv")
odi_bat.head()
```

```python
odi_bowl=pd.read_csv("D:\Practical\Data Sets for Practics\Cricket\Odi_bowl.csv")
odi_bowl.head()
```

```python
odi_team=pd.read_csv("D:\Practical\Data Sets for Practics\Cricket\Odi_team.csv")
odi_team.head()
```

```python
# Sort Runs in Descendin Order

odi_bat.sort_values("Run",ascending=False)
H_run=odi_bat.head(10)
print(H_run)

plt.figure(figsize=(25,10))
sns.barplot(x="Player",y="Run",data=H_run,palette='Dark2')
plt.title("Highest Runs in ODI Cricket",fontsize=30,color="g")
plt.xlabel("Player Name",fontsize=25,color="r")
plt.ylabel("Runs",fontsize=25,color="r")
plt.savefig("H_Run.png",dpi=1000,facecolor="c",bbox_inches="tight")
plt.show()
```

```python
# Sort Wicket in Descending Order

H_wkt=odi_bowl[["Player","Wkts"]].sort_values(["Wkts"],ascending=False)

print(H_wkt)

plt.figure(figsize=(25,10))
sns.barplot(x="Player",y="Wkts",data=H_wkt[0:10],linewidth=2,edgecolor="r",linestyle=":")
plt.title("Highest Wickets in ODI Cricket",fontsize=30,color="orange")
plt.xlabel("Player Name",fontsize=25,color="r")
plt.ylabel("Wicket",fontsize=25,color="r")
plt.savefig("H_Wkts.png",dpi=1000,facecolor="c",bbox_inches="tight")
plt.show()
```

```python
# Sorting Data in Descending Order
team_won=odi_team.sort_values("Won",ascending=False)
team_won
```

```python
# Represent Highest Match Won by Teams

plt.figure(figsize=(20,10))
sns.set_context("paper",font_scale=1.5)

sns.barplot(x=team_won.Team[0:10],y=team_won.Won,palette="viridis_r",linewidth=2,edgecolor="r")

plt.title("Highest Matchs Won by Teams",fontsize=30,color="r")
plt.xlabel("Team Name",fontsize=25,color="orange")
plt.ylabel("Won Matchs",fontsize=25,color="orange")
plt.grid()
plt.savefig("H_won_match.png",dpi=1000,facecolor="c",bbox_inches="tight")
plt.show()
```

```python
# Represent Match Won Percentage by Teams

plt.figure(figsize=(25,10))

plt.pie(list(team_won["%W"]),labels=list(team_won["Team"]),autopct="%0.1f%%",
        radius=1.1,wedgeprops={"width":0.3,"linewidth":2,"edgecolor":"r","linestyle":"--"},
        textprops={"fontsize":12,"color":"brown"})

plt.title("Match Winning Percentage by Teams",fontsize=25,color="g")
plt.savefig("Win_percentage.jpg",dpi=1000,facecolor="c",bbox_inches="tight")
plt.show()
```

```python
# Represent Maech Losing Percentage by Teams
plt.figure(figsize=(25,10))

plt.pie(list(team_won["%L"]),labels=list(team_won["Team"]),autopct="%0.1f%%",startangle=180,
        radius=1.1,wedgeprops={"width":0.3,"linewidth":2,"edgecolor":"b","linestyle":"--"},
        labeldistance=1.1,textprops={"fontsize":12,"color":"brown"})

plt.title("Match Losing Percentage by Teams",fontsize=25,color="r")
plt.savefig("Loss_percentage.jpg",dpi=1000,facecolor="c",bbox_inches="tight")
plt.show()
```

```python
# Find best Bowling Average by Bowlers

b_average=odi_bowl.sort_values("Ave",ascending=True)
b_average.head()
```

```python
# Represent Best Bowling Average by Bowlers
plt.figure(figsize=(25,10))
sns.lineplot(x="Player",y="Ave",data=b_average[0:10],marker="o",
             mfc="r",ms=20,linewidth=2,color="y")
plt.title("Top 10 Best Bowling Average by Players",fontsize=30,color="g")
plt.xlabel("Players Name",fontsize=25,color="r")
plt.ylabel("Averange",fontsize=25,color="r")
plt.grid()
plt.savefig("Best_bowl_ave.png",facecolor="c",dpi=1000,bbox_inches="tight")
plt.show()
```

```python
# Find maximum Century by Player

century=odi_bat[["Player","Century"]].sort_values(["Century"],ascending=False)
century=century.head(10)
```

```python
# Represent Highest Century by Batsmans

plt.figure(figsize=(30,10))
sns.scatterplot(x="Player",y="Century",data=century,linewidth=18,edgecolor="r",
                linestyle="--",palette="Dark2")
plt.title("Highest Century By Batsmans",fontsize=30,color="r")
plt.xlabel("Batsman Name",fontsize=25,color="orange")
plt.ylabel("No. of Century",fontsize=25,color="orange")
plt.grid()
plt.savefig("H_century.png",dpi=1000,facecolor="c",bbox_inches="tight")
plt.show()
```

```python
# Represent Total Catch by Histogram Graph

plt.figure(figsize=(30,10))
sns.set_style("white")
cat=[50,100,150,200,250,300,350,400,450,500]

sns.displot(x=odi_bat["Catch Out"],bins=cat,kde=True,color="m")

plt.title("Total Catch by All Player",fontsize=20,color="g")
plt.xlabel("Catch Out",fontsize=15,color="r")
plt.ylabel("No. of Player",fontsize=15,color="r")
plt.grid()
plt.savefig("Total_catch.png",facecolor="c",dpi=1000,bbox_inches="tight",transparent=True)
plt.show()
```

```python
# Represent Play Total Match by Team

plt.figure(figsize=(50,15))
plt.step(list(odi_team["Team"]),list(odi_team["Mat"]),
         marker="*",ms=35,mfc="y",color="r",linewidth=3,linestyle="--")

plt.title("Played Matches by Teams",fontsize=70,color="m")
plt.xlabel("Teams Name",fontsize=50,color="g")
plt.ylabel("No. of Matches",fontsize=50,color="g")
plt.grid()
plt.savefig("Play_Matches.jpg",dpi=1000,facecolor="c",bbox_inches="tight")
plt.show()
```
